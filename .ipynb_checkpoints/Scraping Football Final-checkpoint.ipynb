# Création de la DB
Code pour recréer la DB en laissant la table Teams inchangée car les liens de scrap permettant de la renseigner ne sont plus valides.

```python
with conn: #PLAYERS OK
    conn.execute("""CREATE TABLE IF NOT EXISTS players (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        first_name VARCHAR,
        last_name VARCHAR,
        position VARCHAR,
        birthdate DATE,
        nationality VARCHAR,
        team_id INTEGER,
        Timestamp DATETIME DEFAULT CURRENT_TIMESTAMP,
        FOREIGN KEY (team_id)
        REFERENCES teams (id)
    );""")

with conn: #CONTRACTS OK
    conn.execute("""CREATE TABLE IF NOT EXISTS contracts (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        player_id INTEGER,
        team_id INTEGER,
        Timestamp DATETIME DEFAULT CURRENT_TIMESTAMP,
        FOREIGN KEY (player_id)
        REFERENCES players (id),
        FOREIGN KEY (team_id)
        REFERENCES teams (id)
    );""")


with conn: #MATCHES OK
    conn.execute("""CREATE TABLE IF NOT EXISTS matches (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        date DATETIME,
        place VARCHAR,
        rainfall FLOAT,
        temperature FLOAT,
        Timestamp DATETIME DEFAULT CURRENT_TIMESTAMP
    );""")

with conn: #MATCHES_TEAM
    conn.execute("""CREATE TABLE IF NOT EXISTS matches_teams (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        match_id INTEGER,
        team_id INTEGER,
        home BOOLEAN,
        team_goal INTEGER,
        points INTEGER,
        Timestamp DATETIME DEFAULT CURRENT_TIMESTAMP,
        FOREIGN KEY (team_id)
        REFERENCES teams (id),
        FOREIGN KEY (match_id)
        REFERENCES matches (id)
    );""")


with conn: #GOALS
    conn.execute("""CREATE TABLE IF NOT EXISTS goals (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        player_id INTEGER,
        match_id INTEGER,
        Timestamp DATETIME DEFAULT CURRENT_TIMESTAMP,
        FOREIGN KEY (player_id)
        REFERENCES players (id),
        FOREIGN KEY (match_id)
        REFERENCES matches (id)
    );""")
```

```python
club_urls = pd.read_csv("clubs_urls.csv")
assert len(club_urls) == 20
# on supprime Dijon dont la page n'est pas structurée de la même manière que les autres.
club_url_dijon = club_urls.iloc[3,0]
club_urls = club_urls.drop(index=3, axis=0)
club_url_dijon
```

## Cas général - sans Dijon

```python
for url in club_urls:
    page = f"https://www.lequipe.fr{url}"
    response = requests.get(page)
    soup = BeautifulSoup(response.content, "html.parser")
    df = pd.read_html(page)
    
    players = df[2]
        
    #first_name = list(players.iloc[:,0])
    # listes :
    last_name = list(players.iloc[:,1])
    position = list(players.iloc[:,4])
    birthdate = list(players.iloc[:,3])
    nationality = list(players.iloc[:,2])
    # string
    team_city = soup.h1.text.split(" ")[0]
    

    
    # Aller chercher dans la table Teams grâce à SQL la city puis l'ID correspondant
    # On le met dans la boucle last_name pour correctement itérer
    # result_2 devient un tuple avec result[0]
    # extend = pareil que append mais pour un tuple
    teams_id = []
    for i in list(last_name):
        curs.execute("SELECT id FROM teams where city like ?", ('%' + team_city + '%',))
        result = curs.fetchall()
        result_2 = result[0]
        teams_id.extend(result_2)

    # total_players sans list = class zip
    # zip n'accepte que des listes
    total_players = zip(last_name, position, birthdate, nationality, teams_id)
    
    # Transformer en liste comprehension
    for i in list(total_players):
        i = list(i)
       
        
        curs.execute('''INSERT INTO players (last_name, position, birthdate, nationality, team_id) VALUES (?, ?, ?, ?, ?);''', (i[0], i[1], i[2], i[3], i[4]))

        conn.commit()
            
```

Itérer sur chaque ligne du tableau, en faisant les actions suivantes :

- j'extrais de ma db l'id correspondant (requête SQL)
- je récupère les id des team domicile et extérieur :

    - je fais une requête sur l'id domicile

```python
# import bibliothèques python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup

# connection à la DB
import sqlite3
from sqlite3 import connect
# conn = connect("ligue_1_final_sauvegarde_bdd_revue_teams.sqlite3")
# curs = conn.cursor()

# import du csv stockant les données json issues du scrap
data = pd.read_csv("matches_teams_bon.csv")
# modification du type de date pour supprimer les HH:MM:SS
data["dates_2"] = data["dates"].apply(lambda x : str(x).split(" ")[0])
data["dates_2"] = pd.to_datetime(data["dates_2"])



for i in range(data.shape[0]):
    
    date = data.loc[i,"dates_2"]
    print(date)
    date = date.to_pydatetime()
    place = data.loc[i, "places"]
    print(place)
    
    # meteo :
    regions = ["aquitaine", "alsace", "ardeche", "auvergne", "bourgogne", "bretagne", "centre", "champagne-ardenne", "corse", "franche-comte", "ile-de-re", "ile-de-france", "languedoc-roussillon", "limousin", "lorraine", "midi-pyrenees", "nord-pas-de-calais", "normandie", "pays-de-la-loire", "picardie", "poitou-charentes", "provence-alpes-c-te-d-azur", "rh-ne-alpes"]
    
    #date_urli = str(date).replace("-","/")
    date_urli = str(date).replace("-","/").split(" ")[0]
    place_urli = place.lower()
    for region in regions:
        page_meteo = f"https://www.historique-meteo.net/france/{region}/{place_urli}/{date_urli}/"
        response = requests.get(page_meteo)
        if response.ok:
            result = pd.read_html(response.text)
            fiche_meteo = result[0]
            temp_max = fiche_meteo.iloc[0,3]
            temp_max = int(temp_max.replace("°",""))
            temp_min = fiche_meteo.iloc[1,3]
            temp_min = int(temp_min.replace("°",""))
            temp = np.mean([temp_min, temp_max])
            precipitations_1 = fiche_meteo.iloc[4,3]
            if not precipitations_1.endswith("mm"):
                precipitations = 0
            else:
                precipitations = int(precipitations_1.replace("mm", ""))
    
    # ajout dans la DB table matches   
    curs.execute("INSERT INTO matches (date, place, rainfall, temperature) VALUES (?, ?, ?, ?);",(date, place, precipitations, temp))
    conn.commit()
    
    # récupération de l'id du match
    curs.execute("SELECT id FROM matches WHERE id=(SELECT max(id) FROM matches)")
    result = curs.fetchone()
    match_id = result[0]    
    
    
    # récupération de l'id de la team domicile
    domicile_name = data.loc[i,"domiciles"]
    curs.execute("SELECT id FROM teams where city like ?", ('%' + domicile_name + '%', ))
    result = curs.fetchone()
    team_domicile_id = result[0]
    
    # récupération de l'id de la team extérieur
    exterieur_name = data.loc[i,"exterieurs"]
    curs.execute("SELECT id FROM teams where city like ?", ('%' + exterieur_name + '%', ))
    result = curs.fetchone()
    team_exterieur_id = result[0]    
    
    # création de la matches_teams domicile
        # STEP 1 : booleen sur domicile
    domicile = True
        # STEP 2 : score en fonction de la colonne vainqueur
    if data.loc[i, "vainqueurs"] == "domicile":
        score_domicile = 3
    elif data.loc[i, "vainqueurs"] == "exterieur":
        score_domicile = 0
    elif data.loc[i, "vainqueurs"] == "ex-aequo":
        score_domicile = 1
    else:
        print("erreur")
        # STEP 3 : identification des buts domicile
    domicile_goals = int(data.loc[i, "scores_domicile"])
        # STEP 4 : rajout des items domicile à la DB
    curs.execute("INSERT INTO matches_teams (match_id, team_id, home, team_goal, points) VALUES (?, ?, ?, ?, ?);",(match_id, team_domicile_id, domicile, domicile_goals, score_domicile))
    conn.commit()
    
    # création de la matches_teams extérieur
        # STEP 1 : booleen sur domicile
    domicile = False
        # STEP 2 : score en fonction de la colonne vainqueur
    if data.loc[i, "vainqueurs"] == "domicile":
        score_exterieur = 0
    elif data.loc[i, "vainqueurs"] == "exterieur":
        score_exterieur = 3
    elif data.loc[i, "vainqueurs"] == "ex-aequo":
        score_exterieur = 1
    else:
        print("erreur")
        # STEP 3 : identification des buts exterieur
    exterieur_goals = int(data.loc[i, "scores_exterieur"])
        # STEP 4 : rajout des items extérieur à la DB
    curs.execute("INSERT INTO matches_teams (match_id, team_id, home, team_goal, points) VALUES (?, ?, ?, ?, ?);",(match_id, team_exterieur_id, domicile, exterieur_goals, score_exterieur))
    conn.commit()
    
    
    # création de la table goals (player_id, match_id):
    
        # STEP 1 : scrap de la page web pour récupérer les noms des joueurs    
    page = data["links"][i] 
    response = requests.get(page)
    soup = BeautifulSoup(response.content, "html.parser")
    fiche_buts_match = soup.find_all("div", "grid__content Scoreboard__goalList")
    for team in fiche_buts_match:
        buts = team.find_all("div", "Scoreboard__goals")
        for but in buts:
            players = but.find_all("span", "Scoreboard__scorer")
            print(players)
            for player in players:
                    name_player = player.text 
                    print(name_player)
                    # faire matcher avec le nom du player dans la table player pour retirer le player_id
                    curs.execute("SELECT id FROM players where last_name like ?", ('%' + name_player + '%', )) 
                    result = curs.fetchone()
                    player_id = result[0]
                    
                    # Step 2 : rajout des items dans la table goals :
                    curs.execute("INSERT INTO goals (player_id, match_id) VALUES (?, ?);",(player_id, match_id))
                    conn.commit()
                    print('--------' *10)
                    
```
